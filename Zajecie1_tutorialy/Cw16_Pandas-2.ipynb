```python
df.groupby('cause').agg('mean') # średnia ze wszystkich kolumn liczbowych dla każdej przyczyny zgonu
```

```python
df_cause = df.groupby('cause').agg({'year': ['count'],
                        'val': ['mean', 'median']})
# statystyki dotyczące przyczyn zgonów
# dla kolumny 'year' count (liczba unikalnych wartości), dla 'val' - średnia i mediana
df_cause
```

```python
df_cause.index # indeksy - przyczyny zgonów
```

```python
df_cause.columns # nazwy kolumn
```

```python
df_cause['val']['mean'].sort_values(ascending = False)
```

```python
df_gender = df.groupby(['cause', 'sex']).agg({'val': ['mean', 'median']})
df_gender
```

```python
df_gender.xs('Transport injuries', level = 'cause') #  wybór według wielu indeksów
```

```python
df_gender.xs('Transport injuries', level = 'cause')['val'].loc["Male",'median'] 
# wybór wartości według wskaźnika multiindex + metoda loc
```

```python
import matplotlib.pyplot as plt # zaimportuj moduł pyplot z biblioteki matplotlib
%matplotlib inline 
# wskazanie, że wykresy należy rysować bezpośrednio w zeszycie, a nie w osobnej zakładce
```

```python
df[(df['location'] == 'Poland') & (df['cause'] == 'HIV/AIDS and sexually transmitted infections') &
  (df['sex'] == 'Both')].pivot_table(values='val', index='year', columns='sex', aggfunc='mean',
                                          fill_value=None, margins=False, dropna=True).plot(kind = 'line')

plt.ylabel('deaths per 100 000 people') # etykieta osi y
plt.title('Deaths from HIV/AIDS and STI in Poland') # tytuł wykresu
```

```python
df_bar = df[(df['sex'].isin(['Male','Female'])) & (df['year'] == 2019)].pivot_table(values='val', 
                index='cause', columns='sex', aggfunc='mean',
                      fill_value=None, margins=False, dropna=True)
df_bar.plot(kind = 'bar')
plt.ylabel('deaths per 100 000 people')
plt.title('deaths by sex')
```
