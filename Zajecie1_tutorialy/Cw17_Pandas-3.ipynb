### Łączenie ramek danych za pomocą metod `merge` i `concat`

```python
df1 = pd.read_csv('IHME-GBD_2019_DATA-15798851-2/IHME-GBD_2019_DATA-15798851-2.csv')

# załaduj drugi plik zapisany na komputerze
# link do pobrania https://drive.google.com/file/d/1uK1edpdqVFJnFaGTWGpVOlaJHgoCBwli/view?usp=sharing 
# źródło danych Global Burden of Disease Study 2019 (GBD 2019) Results.
# Seattle, United States: Institute for Health Metrics and Evaluation (IHME), 2020.
# Strona http://ghdx.healthdata.org/gbd-results-tool
# DALYs (Disability-Adjusted Life Years) — ten wskaźnik oznacza
# ile lat zdrowego i produktywnego życia tracą ludzie z powodu niepełnosprawności i przedwczesnej śmierci z powodu różnych chorób
df1.head(10) # wyświetlamy pierwszy 10 wierszy
```

```python
df2 = pd.read_csv("IHME-GBD_2019_DATA-ff08d9bc-1/IHME-GBD_2019_DATA-ff08d9bc-1.csv")

# załaduj pierwszy plik zapisany na komputerze
# link do pobrania https://drive.google.com/file/d/16fM_JEC5GAkYEDaDgzROrBm8loi9RZRZ/view
# źródło danych Global Burden of Disease Study 2019 (GBD 2019) Results, сайт http://ghdx.healthdata.org/gbd-results-tool
# w pliku dane o zgonach osób z różnych przyczyn dla 204 krajów w latach 2000-2019

df2.head(10)
```

```python
df1.rename(columns = {'val': 'daly_val', 'upper':'daly_upper', 'lower':'daly_lower'}, inplace = True)
df2.rename(columns = {'val': 'deaths_val', 'upper': 'deaths_upper', 'lower':'deaths_lower'}, inplace = True)
# zmień nazwy kolumn zgodnie ze wskaźnikami: liczba utraconych lat produkcyjnych dla 1 ramki danych (DALY)
# i śmiertelność dla drugiej ramki danych ()
```

```python
df_all = pd.merge(df1, df2, on = ['location', 'sex', 'age', 'cause', 'metric', 'year'], how = 'inner') 

# inner: weź tylko dopasowania z dwóch ramek danych
# określiliśmy nazwy kolumn, według których łączymy dataframe'y, ale można też łączyć według indeksu -
# liczba porządkowa, która jest pozostawiona pionowo z [df1.index] bez cudzysłowów
```

```python
 # podziel ramkę danych na dwie ramki danych: pierwsza z wierszami od pierwszego do 50 000, druga od 50 000 do ostatniego
df_all_1 = df_all.iloc[:50000,:]
df_all_2 = df_all.iloc[50000:,:]
```

### 2. Dodawanie nowych kolumn za pomocą operacji matematycznych

```python
# dodaj nową kolumnę, która będzie miała zaokrągloną wartość z innej kolumny
df_all["deaths_val_round"] = df_all["deaths_val"].round(decimals = 1) 
```

```python
df_all
```

```python
# dodać nową kolumnę sumując wartości kilku innych
df_all["deaths_total"] = df_all["deaths_val"] + df_all["deaths_upper"] + df_all["deaths_lower"]
df_all
```

```python
# dodać nową kolumnę sumując wartości kilku innych metodą loc
# oprócz sumy można odczytać mean (średnią), median (medianę), max, min, std (odchylenie standardowe) itp.
df_all["total_loc"] = df_all.loc[:, "deaths_val":"deaths_lower"].sum(axis = 1)
df_all
```

```python
# dodaj nową kolumnę sumując wartości kilku innych metodą iloc
# co ważne, przy korzystaniu z metody iloc nie jest uwzględniana właściwa wartość (numer kolumny), więc jeśli potrzebujemy kolumn z liczbami
# 11, 12 i 13, musisz podać od 11 do 14: df_all.iloc [:, 11:14] 

df_all["total_iloc"] = df_all.iloc[:, 11:14].sum(axis = 1)
```

```python
df_all
```

```python
# usuń niepotrzebne kolumny, aby nie przeszkadzać
df_all.drop(['deaths_val_round','deaths_total', 'total_loc','total_iloc'], axis = 1, inplace = True)
```

### 3. Dodawanie nowych kolumn z pomocą funkcji lambda

```python
# tworzymy listę krajów
CIS_2020 = ['Poland', 'Hungary', 'Italia', 'Germany', 'France', 
            'Spain', 'Romania']
```

```python
# za pomocą funkcji lambda określamy, że jeśli kraj („location”) jest zawarty w liście, to nowa kolumna będzie miała postać „CIS_2020”
# będzie true, jeśli nie, to false
df_all['CIS_2020'] = df_all['location'].apply(lambda x: True if x in CIS_2020 else False )
df_all[df_all['CIS_2020'] == True]
```

```python
# można napisać krócej bez pełnego zapisania tego warunku
df_all['CIS_2020'] = df_all['location'].apply(lambda x: x in CIS_2020)
```

```python
# wybierz wierszy, które pasują do nowego warunku i sprawdź, czy zostały pomyślnie dodane
df_all[df_all['CIS_2020'] == True]
```

```python
# w ten sam sposób możesz zapisać dowolny złożony warunek w osobnej funkcji, a następnie zastosować go do żądanej kolumny
def if_cis(x):
    if x in CIS_2020:
        return True
    else:
        False
```

```python
df_all['CIS_2020'] = df_all['location'].apply(if_cis)
```

```python
df_all[df_all['CIS_2020'] == True]
```

```python
# zapisz ramkę danych do csv
df_all.to_csv('df_all.csv')
```

### 4. Praca z dużymi plikami przy użyciu argumentu chunksize

```python
# podziel duży plik na części za pomocą argumentu chunksize, który określa, ile wierszy powinno znajdować się w każdej części
# wydrukuj 5 wierszy każdej wynikowej części

for chunk_df in pd.read_csv('df_all.csv', 
                      chunksize = 50000):
    print("CHUNK DF")
    print(chunk_df.head())
```

```python
# zastosuj metodę groupby oddzielnie do każdej części, a następnie połącz wynik w nową ramkę danych
new_df = pd.DataFrame() # pusta ramka danych
for chunk_df in pd.read_csv('df_all.csv', 
                      chunksize = 50000):
    result = chunk_df.groupby(['cause', 'year']).agg({'deaths_val': 'mean',
                                                    'daly_val': 'mean'})
    new_df = pd.concat([new_df,result])
```

```python
new_df
```

```python
# zapisz wynik do pliku
new_df.to_csv('deaths_daly_groupby.csv', encoding = 'utf-8')
```
